```python
pltdos.SHOW_TOTAL_DOS  = True
pltdos.SHOW_FERMI_LINE = True
pltdos.ALIGN_TO_FERMI = True
pltdos.X_LIM           = [-5, 5]
pltdos.Y_LIM = None
pltdos.GENERAL_FONT_SIZE = 14
pltdos.LEGEND_FONT_SIZE = 8

# ===== 刻度设置（对应 old plotter 的 X/Y_TICK_INTERVAL）=====
pltdos.ENABLE_CUSTOM_TICKS = True
pltdos.X_TICK_INTERVAL = 1.0
pltdos.Y_TICK_INTERVAL = 2.0  # 需要就设，不需要就 None

# ===== 配色设置（对应 old plotter 的元素/轨道映射）=====
pltdos.COLOR_MODE = "mapped"
pltdos.ELEMENT_COLORS = {
    "Sc": "#5FBAAF",
    "Cu": "#C29FD0",
}
pltdos.ORBITAL_COLORS = {
    "dxy": "#8ECFB0",
    "dxz": "#AFB6D2",
    "dyz": "#D8B5AF",
    "dz2": "#A5C1DE",
    "dx2y2": "#E2D185",
}

# 可选：同轨道不同元素不同色（新能力）
# pltdos.ELEMENT_ORBITAL_COLORS = {
#     "V": {"dxy": "#1f77b4", "dx2-y2": "#ff7f0e"},
#     "Ti": {"dxy": "#2ca02c", "dx2y2": "#d62728"},
# }

pltdos.LEGEND_LOC      = 'upper right'
pltdos.ORDER_MODE      = 'manual'
pltdos.MANUAL_ORDER    = ['Fe', 'Ti', 'O', 'N', 'Cl']

# ===== 自动保存 =====
pltdos.AUTO_SAVE_FIGURE = True
pltdos.SAVE_FIGURE_PATH = "dos.png"
pltdos.SAVE_FIGURE_DPI = 600  # 你当前要求的默认值

# ===== 出图 =====
dos_data = calc.dos.to_dict(selection="d(Sc), d(Cu), p(O)")
pltdos.plot_vasp_dos(dos_data)
```
